### Hypothesis Testing

#### A. Older individuals may not possess the same level of technological proficiency and may, therefore, lean toward traditional in-store shopping preferences.

* Null Hypothesis (Hâ‚€): Older individuals (Age >50) and yonger individuals (Age <= 50) have no difference in their preference for in store shoping.

* Alternative Hypothesis (Hâ‚): Older individuals (Age > 50) likly to prefer traditional in-store shopping more than yonger individuals.

```python
t_stats, p_value = stats.ttest_ind(Marketing_Camp[Marketing_Camp["Age"]>50]['NumStorePurchases'],Marketing_Camp[Marketing_Camp["Age"]<=50]['NumStorePurchases'])
print(f"T-statistic: {t_stat}, P-value: {p_value}")
if p_value < 0.05:  # 5% significance level
    print("Reject the null hypothesis: Age and shopping preferences are associated.")
else:
    print("Fail to reject the null hypothesis: No significant association.")
```

 #### B.Customers with children likely experience time constraints, making online shopping a more convenient option.

* Null Hypothesis (Hâ‚€): Having children does not significantly impact the likelihood of choosing online shopping as a preferred option. There is no association between having children and shopping preferences.

* Alternative Hypothesis (Hâ‚): Having children increases the likelihood of choosing online shopping as a preferred option. There is a significant association between having children and shopping preferences.


```python
t_stat, p_value = stats.ttest_ind(marketing_data[marketing_data['Total_Children'] > 0]['NumWebPurchases'], marketing_data[marketing_data['Total_Children'] == 0]['NumWebPurchases'])
print(f"T-statistic: {t_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject H0: Customers with children significantly prefer online shopping.")
else:
    print("Fail to reject H0: No significant preference difference.")
```

#### C. Sales at physical stores may face the risk of cannibalization by alternative distribution channels. 

* Null Hypothesis (Hâ‚€): Sales at physical stores are not significantly affected by alternative distribution channels. There is no cannibalization effect.

* Alternative Hypothesis (Hâ‚): Sales at physical stores are significantly reduced due to cannibalization by alternative distribution channels.

```python
t_stat, p_value = stats.ttest_ind(marketing_data['NumStorePurchases'], marketing_data[['NumWebPurchases', 'NumCatalogPurchases']].sum(axis=1))
print(f"T-statistic: {t_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject H0: Store sales are significantly affected by other channels.")
else:
    print("Fail to reject H0: No significant impact of other channels on store sales.")
```

#### D.Does the United States significantly outperform the rest of the world in total purchase volumes? 

*  Null Hypothesis (H0): The total purchase volumes in the US and other countries are the same.
* Alternative Hypothesis (H1): The total purchase volume in the US is significantly different from other countries.

```python
t_stat, p_value = stats.ttest_ind(marketing_data[marketing_data['Country'] == 'US']['Total_Purchases'], marketing_data[marketing_data['Country'] != 'US']['Total_Purchases'])
print(f"T-statistic: {t_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject H0: The US has significantly different purchase volumes.")
else:
    print("Fail to reject H0: No significant difference in purchase volumes.")
```
